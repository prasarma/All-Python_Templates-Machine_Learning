```python

#DATA IMPORT
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
titanic.age.fillna(0,inplace=True)
```

```python

                                                # MATPLOTLIB 
```

```python
# MATPLOTLIB UNIVARIATE
```

```python
# Checking the value count of a column value
titanic['survived'].value_counts()
titanic.survived.value_counts()
```

```python
# Checking the value count of a column value percentage wise
titanic['survived'].value_counts(normalize=True)
titanic.survived.value_counts(normalize=True)
```

```python
# Creating plot bar for value count of a column  (value wise)
titanic['survived'].value_counts().plot.bar()
```

```python

# Creating plot bar for value count of a column  (percentage wise)
titanic['survived'].value_counts(normalize=True).plot.bar()
```

```python

# Let do some plotting
#train.ftypes
department_count = titanic["sex"].value_counts()
plt.figure(figsize=(14,5))
plt.bar(department_count.index, department_count.values)
plt.xlabel("survived")
plt.ylabel("Frequency")
plt.title("survived")
plt.show()
```

```python
#Creating graph for categorical values (4 Bar plots in one screen)
#print titanic
plt.figure(1)

plt.subplot(221)
titanic['sex'].value_counts(normalize=True).plot.bar(title='sex')

plt.subplot(222)
titanic['pclass'].value_counts(normalize=True).plot.bar(title='pclass')

plt.subplot(223)
titanic['class'].value_counts(normalize=True).plot.bar(title='class')

plt.subplot(224)
titanic['sibsp'].value_counts(normalize=True).plot.bar(title='sibsp')

plt.show()
```

```python

# Creating histogram of Item Visibility
titanic['age'].hist(bins=10)
```

```python
# MATPLOTLIB BI-VARIATE
```

```python
# Value count using groupby
titanic.groupby('sex').survived.value_counts(normalize=True)
```

```python
# Creating bar plot of value count using groupby
titanic.groupby('sex').survived.value_counts().plot.bar()
```

```python
# Creating bar plot percentage wise using groupby   
titanic.groupby('sex').survived.value_counts(normalize=True).plot.bar()
```

```python

# Creating two tabs one on another(categorical vs Target(categorical))         PERCENTAGE
Gender_lnst=pd.crosstab(titanic.sex,titanic['survived'])
Gender_lnst.div(Gender_lnst.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('sex')
plt.ylabel('Percentage')
plt.show()
```

```python

# Creating two tabs one on another(categorical vs Target(categorical))         VALUE COUNT
sex_count = titanic.groupby(["sex", "survived"])["survived"].count()
sex_count.unstack().plot(kind='bar', stacked=True)
```

```python

#(BINNING) Creating Numerical column into bins (convert to categorical)vs Target (categorical)
bins=[0,20,25,30,40]
group=['Low','Average','High', 'Very high']
titanic['age_bin']=pd.cut(titanic['age'],bins,labels=group)
Income_bin=pd.crosstab(titanic['age_bin'],titanic['survived'])
Income_bin.div(Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('ApplicantIncome')
plt.ylabel('Percentage')
```

```python
''' 
1. single categorical 
    1.1 count plot ==> countplot()

2. Two categorical (categorical vs categorical)
    2.1 countplot ==> countplot()
    2.2 barplot  ==> barplot()
    
3. single categorical and one continuous (categorical vs continuous)
    3.1 scatter plot  ==> catplot  (default)
    3.2 swarm plot ==> catplot (kind=swarm)
    3.3 box plot ==> catplot (kind=box)
    3.4 violin plot ==> catplot (kind=violin)
    3.5 point plot ==>  catplot (kind=point )
    
4. single continuous 
    4.1 distribution plot ==> distplot()
    4.2 distribution plot (probability density function)==> distplot (hist=False)

5. Two continuous (continuous vs continuous) and one single categorical (continuous vs continuous vs categorical)
    5.1 scatter plot (2 continuos)==> relplot(default)
    5.2 scatter plot (2 continuos and 1 categorical variable)==> relplot(default)
    5.3 line plot (2 continuos)==> relplot(kind=line)
    5.4 line plot (2 continuos and 1 categorical variable)==> relplot(kind=line)

6. regression plot
    6.1 regression plot (between two variables only) ==> regplot()
    6.2 regression plot (between three variables) ==> lmplot()


'''
```

```python
'''1. single categorical 
    1.1 count plot ==> countplot()      '''

sns.countplot(x='sex',data=titanic)
```

```python
'''1. single categorical 
    1.1 count plot ==> countplot()    horizontal  '''

sns.countplot(y='sex',data=titanic)
```

```python
'''2. Two categorical (categorical vs categorical)
    2.1 countplot ==> countplot()'''

sns.countplot(x='sex',hue='survived',data=titanic)
```

```python
'''2. Two categorical (categorical vs categorical)
    2.1 countplot ==> countplot()          horizontal '''

sns.countplot(y='sex',hue='survived',data=titanic)
```

```python
'''2. Two categorical (categorical vs categorical)
    2.2 barplot ==> barplot()'''
sns.barplot(x='sex',y='survived',data=titanic,ci=None)
```

```python
'''2. Two categorical (categorical vs categorical)
    2.2 barplot ==> barplot()        horizontal    '''
sns.barplot(y='sex',x='survived',data=titanic,ci=None)
```

```python
'''3. single categorical and one continuous (categorical vs continuous)
    3.1 scatter plot  ==> catplot  (default)'''
sns.catplot(x='survived',y='age',data=titanic)
```

```python
'''3. single categorical and one continuous (categorical vs continuous) and one more categorical variable in hue
    3.1 scatter plot  ==> catplot  (default)        '''
sns.catplot(x='survived',y='age',hue='sex',data=titanic)
```

```python
'''3. single categorical and one continuous (categorical vs continuous)
    3.2 swarm plot  ==> catplot  (kind=swarm)'''
sns.catplot(x='survived',y='age',data=titanic,kind='swarm')
```

```python
'''3. single categorical and one continuous (categorical vs continuous) and one categorical in hue
    3.2 swarm plot  ==> catplot  (kind=swarm)'''
sns.catplot(x='survived',y='age',hue='sex',data=titanic,kind='swarm')
```

```python
'''3. single categorical and one continuous (categorical vs continuous)
    3.3 box plot  ==> catplot  (kind=box)'''
sns.catplot(x='survived',y='age',data=titanic,kind='box')
```

```python
'''3. single categorical and one continuous (categorical vs continuous) and one more column hue
    3.3 box plot  ==> catplot  (kind=box)'''
sns.catplot(x='survived',y='age',hue='sex',data=titanic,kind='box')
```

```python
'''3. single categorical and one continuous (categorical vs continuous)
    3.4 violin plot  ==> catplot  (kind = violin)'''
sns.catplot(x='survived',y='age',data=titanic,kind='violin')
```

```python
'''3. single categorical and one continuous (categorical vs continuous) and one more categorical in hue
    3.4 violin plot  ==> catplot  (kind = violin)'''
sns.catplot(x='survived',y='age',hue='sex',data=titanic,kind='violin')
```
